# Recordings generation

This process requires a set of steps which are:

1. Templates generation
2. Recordings generation

## Templates generation

This generation uses a set of cell models which are the provided by the MEArec package:
https://github.com/alejoe91/MEArec/tree/master/MEArec/cell_models/bbp 

```python
%load_ext autoreload
%autoreload 2
```

```python
config = {
    "config_folder": "../data/mearec",
    "recordings_folder": "/mnt/Archivos/data/output/recordings",
    "templates_folder": "/mnt/Archivos/data/output/templates"
}
```

```python
from eapprocessor.mearec.api import generate_templates, generate_recordings, load_recordings
```

```python
tempgen = generate_templates(config_folder=config["config_folder"],
                            templates_folder=config["templates_folder"],
                            verbose=True, recompile=True)
```

Once generated in a custom location, then it follows to recording generation

## Recordings generation

For the recordings generation, it will be considered a set of noise_levels to perform the evaluation for different values of noise.

```python
noise_level = [5, 10, 15, 20, 25, 50, 75]
```

```python
for noise in noise_level:
     generate_recordings(config_folder=config["config_folder"],
                        templates_folder=config["templates_folder"],
                        recordings_folder=config["recordings_folder"],
                        verbose=True, noise_level=noise)
```

Once generated, the function `loadRecordings` could be used to load the recordings and plot them

```python
recgen_list = [load_recordings(datafolder=config["recordings_folder"],
                         verbose=True, noise_level=noise) for noise in noise_level]
```

```python
%matplotlib inline
```

```python
import MEArec as mr
```

```python
for recgen in recgen_list:
    mr.plot_recordings(recgen=recgen)
    current_noise_level = recgen.info["recordings"]["noise_level"]
    print(f"Noise level={current_noise_level}")
```

```python
for recgen in recgen_list:
    mr.plot_rasters(recgen.spiketrains)
    current_noise_level = recgen.info["recordings"]["noise_level"]
    print(f"Noise level={current_noise_level}")
```

```python
from eapprocessor.plot.signals import plot_transient_recordings
```

```python
for recgen in recgen_list:
    plot_transient_recordings(recgen, channels=range(1))
```

## Recordings generation for different frequencies

```python
fs_list = [20000, 16000]
noise_level = [5, 25, 50, 100, 200, 300]
```

```python
for fs in fs_list:
    for noise in noise_level:
         generate_recordings(config_folder=config["config_folder"],
                            templates_folder=config["templates_folder"],
                            recordings_folder=config["recordings_folder"],
                            verbose=True, noise_level=noise, fs=fs)
```

```python
all_recgen_list = [[load_recordings(datafolder=config["recordings_folder"],
                verbose=True, noise_level=noise, fs=fs) for noise in noise_level]
               for fs in fs_list]
```

Plotting for 16kHz sampling frecuency:

```python
for recgen in all_recgen_list[-1]:
    plot_transient_recordings(recgen, channels=range(1))
```

```python
from eapprocessor.plot.signals import plot_recordings_list
```

```python
recgen_noise_5uv = [ recgen_fs[0] for recgen_fs in all_recgen_list]
```

```python
plot_recordings_list(recgen_noise_5uv, t_stop=0.1, channels=range(1),marker='.--')
```

```python
plot_recordings_list(recgen_noise_5uv, t_start=0.02, t_stop=0.0375, channels=range(1),marker='.--')
```

These variations exists because of the recordings generation which resample the template sources to the new sampling frecuency, it will be different as it will be resampled after the signal is generated.
