```python
from eapprocessor.tools.load import loadNEO, loadCountEvaluation
```

```python
noise_levels = [5, 10, 15, 20, 25, 50 ,75]
resolution = 12
```

```python
neo_list=[]
neo_files=[]
for noise_level in noise_levels:
    neo_dict, file = loadNEO(f"{BASE_DIR}/output/preprocessor", verbose=True, resolution=resolution, 
                       noise_level=noise_level)
    neo_list += [neo_dict]
    neo_files += [file]
```

```python
test_neo = neo_list[0]
```

```python
from eapprocessor.plot.signals import plotSignals
```

```python
plotSignals(test_neo,channels=range(1),crange=range(5000))
```

```python
from eapprocessor.tools.distances import calcDistancesFromRecordings, getOrderByChannels, \
    convertDistancesToIntensity
```

```python
distances = calcDistancesFromRecordings(test_neo["recordings"])
print(distances)
print(distances.shape)
```

```python
orders = getOrderByChannels(distances)
orders
```

```python
amplitudes = convertDistancesToIntensity(distances)
amplitudes
```

```python
plotSignals(test_neo,channels=range(1),crange=range(5000),intensities=amplitudes[0])
```

```python
plotSignals(test_neo,channels=range(1),crange=range(5000),intensities=amplitudes[0], range_spikes=orders[0][:2])
```
