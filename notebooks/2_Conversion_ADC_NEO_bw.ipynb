# Recordings conversion to digital samples and preprocessed by NEO

After the recordings generation, follows the conversion by a modeled ADC and their preprocessing by a NEO module, also modeled with different widths.

## ADC conversion

```python
%load_ext autoreload
%autoreload 2
```

```python
BASE_DIR="/mnt/Archivos/data"
```

```python
from eapprocessor.process import get_converted_adc, get_neo, get_over_threshold
from eapprocessor.tools.load import load_neo
```

First, define the parameters related to the noise level and the resolution and voltage reference. In this first case, the resolution will be 12 and the reference voltage will be 500 as the maximum value for the generated signal.

### Conversion for 20kHz, 12-6bit, 2bw

```python
fs = 20000
noise_levels = [5, 25, 50, 100]
#resolutions = [12,10,9,8,7,6]
resolutions = [5,4,3,2,1]
recfolders = [
    f"{BASE_DIR}/output/recordings",
    f"{BASE_DIR}/output-bw-300-3000/recordings",
    f"{BASE_DIR}/output-bw-300-1000/recordings"
]
voltage_ref = 500
```

```python
for recfolder in recfolders:    
    for resolution in resolutions:    
        for noise_level in noise_levels:
            adcgen_list = get_converted_adc(recfile=recfolder,voltage_ref=voltage_ref, 
                                            resolution=resolution, noise_level = noise_level, fs=fs, verbose=True)
```

### LCADC conversion

```python
from eapprocessor.process import get_converted_adc
```

```python
for noise_level in noise_levels:
    adcgen_list = get_converted_adc(recfile=f"{BASE_DIR}/output/recordings",voltage_ref=voltage_ref, 
                                    resolution=resolution, noise_level = noise_level, fs=fs, is_lcadc=True,
                                    verbose=True)
```

### LCADC at 8bit resolution

## NEO application

Then the NEO preprocessor could be applied. In this case, the formula:
$$ y[n] = x[n]^2 - x[n-w]x[n+w]$$
is applied with different widths $w$.

### NEO application for 20kHz, 12-6bit, 2bw

```python
w = [1,4,16,32]
adcfolders = [
    f"{BASE_DIR}/output/adc",
    f"{BASE_DIR}/output-bw-300-3000/adc",
    f"{BASE_DIR}/output-bw-300-1000/adc"
]
```

```python
for adcfolder in adcfolders:    
    for resolution in resolutions:    
        for noise_level in noise_levels:
            get_neo(adcfile=adcfolder, verbose=True, resolution=resolution, 
                    noise_level=noise_level,w=w, fs=fs)
```

### NEO for LCADC 20kHz 10-6bits

## Threshold evaluation

Then, an evaluation could be performed to every signal: recordings, ADC and NEO, with various values of threshold ranging from $0.1A_{\max}$ to $A_{\max}$.
To speed up the process, only 2 channels (electrodes) are considered: 0 and 1.

### Evaluation for 20kHz 10-6bits, 2bw

```python
channel_indexes=[21,27]
noise_levels=[5,25]
resolutions = [5,4,3]
nthresholds = 50
neofolders = [
    f"{BASE_DIR}/output/preprocessor",
    f"{BASE_DIR}/output-bw-300-3000/preprocessor",
    f"{BASE_DIR}/output-bw-300-1000/preprocessor"
]
```

```python
for neofolder in neofolders:
    for resolution in resolutions:    
        for noise_level in noise_levels:
            get_over_threshold(neofolder, verbose=True, resolution=resolution, 
                             ch_indexes=channel_indexes, 
                             nthresholds=nthresholds, 
                             noise_level=noise_level,
                             fs=fs)
```

### Evaluation for LCADC

```python
for noise_level in noise_levels:
    get_over_threshold(f'{BASE_DIR}/output/preprocessor_lcadc', verbose=True, resolution=resolution, 
                     ch_indexes=channel_indexes, 
                     nthresholds=nthresholds, 
                     noise_level=noise_level,
                     fs=fs, is_lcadc=True)
```

### Evaluation for LCADC 10-6bit

```python
for resolution in resolutions:  
    for noise_level in noise_levels:
        get_over_threshold(f'{BASE_DIR}/output/preprocessor_lcadc', verbose=True, resolution=resolution, 
                         ch_indexes=channel_indexes, 
                         nthresholds=nthresholds, 
                         noise_level=noise_level,
                         fs=fs, is_lcadc=True)
```
